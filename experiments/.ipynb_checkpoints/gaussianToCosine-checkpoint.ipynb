```python
x = np.linspace(-np.pi, np.pi)
cos = lambda x: 1/(2*np.pi)*(1 + np.cos(x))
gaus = lambda x,sigma: 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5 * (x/sigma)**2)

g,c = gaus(x,1.14),cos(x)
plt.plot(x,g, label="Gaussian")
plt.plot(x,c, '--', dashes=(10, 3), label="cosine")
plt.xticks([])
plt.yticks([])
plt.xlim([-3,3])
plt.xlabel(r"wavenumbers")
plt.ylabel("intensity")
plt.legend()
plt.show()
```

```python
img, wavenumbers = data[0][0], data[1][0]
img = img.reshape(-1, img.shape[-1])
```

```python
def median(values):
    return sorted(values)[len(values)//2]
```

```python
threshold1 = lambda z : median(z) - np.min(z)
threshold2 = lambda z : np.mean(z) - np.min(z)
threshold3 = lambda z : np.max(z) - median(z)
threshold4 = lambda z : np.max(z) - np.mean(z)
threshold5 = lambda z : (np.max(z) - np.min(z))/2


labels = ["raw spectrum", "median-min", "mean-min", "max-median", "max-mean"]
threshold = [threshold1, threshold2, threshold3, threshold4]   
```

```python
PLOT = True

plt.rcParams['figure.figsize'] = (10.0, 3.0)
plt.rcParams['figure.dpi'] = 250
plt.rcParams['lines.linewidth'] = 0.8

width = 20

mean_lst = []
std_lst = []
for i, s in enumerate(img):
    if PLOT and i < 12:
        continue
    if PLOT and i > 12:
        break
    t_lst = [[] for _ in threshold]
    for i in range(img.shape[1]-width):
        for ind,t in enumerate(threshold):
            t_lst[ind].append(t(s[i:i+width]))
    if PLOT:
        plt.plot(wavenumbers[width//2:-width//2], s[width//2:-width//2])
        for t in t_lst:
            plt.plot(wavenumbers[width//2:-width//2], t)
        plt.legend(labels)
        plt.ylim(0)
        plt.xlim(wavenumbers[width//2], wavenumbers[-width//2])
        plt.xlabel(r"wavenumbers ($cm^{-1}$)")
        plt.ylabel(r"intensity (-)")
        plt.show()
    
    mean_lst.append(np.mean(t_lst, 1))
    std_lst.append(np.std(t_lst, 1))

wavenumbers[20] - wavenumbers[0]
```

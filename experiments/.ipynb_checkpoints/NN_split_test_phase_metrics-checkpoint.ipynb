```python
plt.rcParams['figure.figsize'] = (12.0, 3.0)
plt.rcParams['figure.dpi'] = 200
plt.rcParams['lines.linewidth'] = 1

epochs = 30
smoothing = 2696 // 8
smoothing = 337
smoothing = 4*30
# smoothing = 1

def avg_error(data):
    error_Our = copy.copy(data)

    error_Our = np.array(error_Our).reshape(-1, smoothing)
    return np.mean(error_Our, 1)

error_name = "compared_grad"
error_name = "MSE raman"
error_name = "train_loss"

error_label = error_name.replace("_", " ")
error_Our = avg_error(data_Our_raw[error_name])
plt.plot(range(len(error_Our)), error_Our, label=f"complete")
error_Our = avg_error(data_Our_RFFT[error_name])
plt.plot(range(len(error_Our)), error_Our, label=f"no RFFT")
error_Our = avg_error(data_Our_phase3[error_name])
plt.plot(range(len(error_Our)), error_Our, label=f"no phase3")

    
    
plt.xticks(ticks=np.linspace(0,len(error_Our), epochs+1), labels=range(0,epochs+1))
plt.xlim(0,len(error_Our)-1)
# plt.ylim(min(error_ALS)-10,max(error_ALS))
# plt.ylim(min(error_ALS)*0.9,max(error_ALS))
# plt.ylim(0,7)
plt.yscale("log")
plt.xlabel("epochs")
plt.ylabel(f"{error_name}")
plt.ylabel(f"{error_label}")
plt.legend()
plt.show()
```

```python
from load_modules import *
import copy 

plt.rcParams['figure.figsize'] = (14.0, 3.0)
plt.rcParams['figure.dpi'] = 200
plt.rcParams['lines.linewidth'] = 1
```

```python
x = np.linspace(0,1,101)
y = np.linspace(0,4,400)
s = np.array([1])
# LSQ.gaussian(x, np.array([0]), s)

B0 = lambda x: 1/6 * (1 -3*x + 3*x**2 - x**3) 
B1 = lambda x: 1/6 * (4 - 6 * x**2 + 3 * x**3) 
B2 = lambda x: 1/6 * (1 + 3 * x + 3 * x**2 - 3*x**3) 
B3 = lambda x: 1/6 * (x**3) 

B = np.concatenate((B3(x)[:-1], B2(x)[:-1], B1(x)[:-1], B0(x)[:-1]))

B1 = copy.copy(B)
B2 = copy.copy(B)
B3 = copy.copy(B)
B4 = copy.copy(B)

T = 1-B1
T[100:] -= B2[:-100]
T[200:] -= B3[:-200]
T[300:] -= B4[:-300]

T2 = 1-B4
T2[100:] -= B3[:-100]
T2[200:] -= B2[:-200]
T2[300:] -= B1[:-300]

plt.plot(y-1,T, label=r"$B_0$")
plt.plot(y-1,B1, label=r"$B_1$")
plt.plot(y,B2, label=r"$B_2$")
plt.plot(y+1,B3, label=r"$B_3$")
plt.plot(y+2,B4, label=r"$B_4$")
plt.plot(y+2,T2[::-1], label=r"$B_5$")
plt.ylim(0)
plt.xlim(0,5)
plt.xlabel(r"knots ($P_i$)")
plt.legend()

```
